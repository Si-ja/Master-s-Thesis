```python
#Sin wave analysis
import numpy as np
import pandas as pd
from pylab import rcParams
import statsmodels.api as sm
import matplotlib.pyplot as plt
import torch
```

```python
seed_value = 42
INSTANCES = 7000
DURATION = 100
ADD_ERR = 0.15

torch.manual_seed(seed_value)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
np.random.seed(seed_value)

x = np.linspace(0, DURATION, INSTANCES)
noise = np.random.normal(0, ADD_ERR, len(x))
y = np.sin(x) + noise

f = plt.figure(figsize=(8, 6), dpi=80)
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
MEDIUM_SIZE = 28
plt.rc('font', size = MEDIUM_SIZE)      
plt.rc('axes', titlesize = MEDIUM_SIZE)     
plt.rc('axes', labelsize = MEDIUM_SIZE)    
plt.rc('xtick', labelsize = MEDIUM_SIZE)    
plt.rc('ytick', labelsize = MEDIUM_SIZE)    
plt.rc('legend', fontsize = MEDIUM_SIZE)    
plt.rc('figure', titlesize = MEDIUM_SIZE)
plt.plot(x, y)
plt.xlabel("Time", fontsize=38)
plt.ylabel("sin deviation", fontsize=38)
plt.title("sin wave with noise", fontsize=38)
f.savefig("Sin_Wave.pdf", bbox_inches='tight')
plt.show()
```

```python
data = pd.Series(y)
MEDIUM_SIZE = 30
plt.rc('font', size = MEDIUM_SIZE)      
plt.rc('axes', titlesize = MEDIUM_SIZE)     
plt.rc('axes', labelsize = MEDIUM_SIZE)    
plt.rc('xtick', labelsize = MEDIUM_SIZE)    
plt.rc('ytick', labelsize = MEDIUM_SIZE)    
plt.rc('legend', fontsize = MEDIUM_SIZE)    
plt.rc('figure', titlesize = MEDIUM_SIZE)
data.plot(figsize=(15, 5))
plt.show()

decomposed = sm.tsa.seasonal_decompose(data.values, freq=438, model='additive')
f = decomposed.plot()
plt.rcParams['figure.figsize'] = (9, 9)
# f.savefig("Seasonal Decomposition.pdf")
f.savefig("Wave_seasonal.pdf", bbox_inches='tight')
plt.show()
```
